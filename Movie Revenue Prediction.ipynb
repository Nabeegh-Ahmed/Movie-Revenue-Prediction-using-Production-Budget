# Predicting Revenue using Production Budget

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
data = pd.read_csv('source/cost_revenue.csv')
```

```python
productionBudget = pd.DataFrame(data, columns=['production_budget_usd'])
worldwideGross   = pd.DataFrame(data, columns=['worldwide_gross_usd'])
```

```python
plt.figure(figsize=(10,6))
plt.scatter(productionBudget, worldwideGross, alpha=0.3)
plt.xlim(0, 450000000)
plt.ylim(0, 3000000000)
plt.title('Production Budget vs Worldwide Gross')
plt.xlabel('Production Budget')
plt.ylabel('Worldwide Gross')
plt.show()
```

```python
regression = LinearRegression()
regression.fit(productionBudget, worldwideGross)

RegressionCoefficient = regression.coef_ # Slope Coefficient
RegressionIntercept   = regression.intercept_ # Slope Intercept

print("Y = ", RegressionCoefficient[0][0], " + ", RegressionIntercept[0], "x")
```

```python
plt.figure(figsize=(10,6))
plt.scatter(productionBudget, worldwideGross, alpha=0.3)
plt.plot(productionBudget, regression.predict(productionBudget), color='red', linewidth=4) # Regression Line
plt.xlim(0, 450000000)
plt.ylim(0, 3000000000)
plt.title('Production Budget vs Worldwide Gross')
plt.xlabel('Production Budget')
plt.ylabel('Worldwide Gross')
plt.show()
```

```python
regression.score(productionBudget, worldwideGross)
```
